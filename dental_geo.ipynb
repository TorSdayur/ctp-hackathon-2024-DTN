```python
import pandas as pd

df = pd.read_csv('csv/dental.csv')
df
```

```python
from geopy.geocoders import Nominatim

# Geocoding function
def geocode(address):
    geolocator = Nominatim(user_agent="my_app")
    location = geolocator.geocode(address)
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

# Combine Address, City, State, Zip into a single string for geocoding
df['Full Address'] = df['Address'] + ', ' + df['City, State, Zip']

# Apply the geocoding function to the DataFrame
df[['Latitude', 'Longitude']] = df['Full Address'].apply(geocode).apply(pd.Series)

df
```
